```python
prices_master_list = []
year_prices_dictionary = {}
month_prices_dictionary = {}

#@title read_gas_prices()  
# Reading from the file and building the prices_master_list
'''
Method Name: read_gas_prices
Input: file name
Returns: None
What global variables are updated? prices_master_list


This function takes in a gas prices file name and creates a prices_master_list
'''
def read_gas_prices(file_name = 'gasprices.txt'):
  # open the file 
  # read the line
  # split the line 
  # do any necessary data conversions 
  # build the mini list called price_mini_list
  # append this price_mini_list to the master list prices_master_list
  # return (None) out of the function
  file_contents = open(file_name, 'r')
  file_lines = file_contents.readlines()


  for s in file_lines:
    m_d_y_list = s.split('-')
    y_p_list = m_d_y_list[2].split(':')

    # Once we split by :, we should be getting exactly two tokens.
    # Otherwise, ignore the line
    if (len(y_p_list)!=2):
      continue

    price = y_p_list[1].strip('\n')

    m_d_y_p_list = []
    month = int(m_d_y_list[0])
    day = int(m_d_y_list[1])
    year = int(y_p_list[0])
    price = float(price)

    m_d_y_p_list.append(month)
    m_d_y_p_list.append(day)
    m_d_y_p_list.append(year)
    m_d_y_p_list.append(price)

    #update the prices master list
    # only if the m_d_y_p_list doesn't exist in the master list
    # this helps us to avoid the duplicates
    if m_d_y_p_list in prices_master_list:
      continue

    # if it is not in the master list,
    # we will continue updating our data structures
    prices_master_list.append(m_d_y_p_list)

    # update the year dictionary
    # Update year_prices_dictionary
    if year not in year_prices_dictionary:
        year_prices_dictionary[year] = []
    year_prices_dictionary[year].append(price)

    # update the month dictionary
    # Update month_prices_dictionary
    if month not in month_prices_dictionary:
        month_prices_dictionary[month] = []
    month_prices_dictionary[month].append(price)


# Testing this fuction by itself
read_gas_prices()


if VERBOSE:
    #If the above method works correctly, then prices_master_list would have been updated with all the data from the file.
    print('Number of mini lists in prices_master_list: ', len(prices_master_list))
    for index in range(5):
      mini_list = prices_master_list[index]
      print(mini_list)

    print('Length of Year Prices Dictionary: ', len(year_prices_dictionary))
    print('Year Prices Dictionary: ', year_prices_dictionary)

    print('Length of Month Prices Dictionary: ', len(month_prices_dictionary))
    print('Month Prices Dictionary: ', month_prices_dictionary)
else:
    print("gasprices.txt data is processed")
```

```python
#@title create_price_range_column()
''' create a text file called "gas_prices_bin_data' with an additional column 
mentioning the bin name for each line item 

date:price:bin_name
'''

range1 = 0.949
range2 = 1.753
range3 = 2.557
range4 = 3.364  # anything above range4 is considered as a "Very High"
range5 = 4.165  

def get_bin_name(input_price):
  bin_name = ' '
  
  if input_price >= range1 and input_price < range2:
    bin_name='Low'
  if input_price >= range2 and input_price < range3:
    bin_name='Medium'
  if input_price >= range3 and input_price < range4:
    bin_name='High'
  if input_price >= range4:
    bin_name='Very High'

  return bin_name

def create_price_range_column():
  with open("gasprices.txt",'r') as data_file:
    with open('gas_prices_binning.txt','w') as new_file:
      for line in data_file:
        gasList = line.split(':')
        date = gasList[0]
        price_str = gasList[1]
        price_str = price_str[:-1]
        price_float = float(price_str)
        pricerange = get_bin_name(price_float)
        new_file.write(date+':'+price_str+':'+pricerange+'\n')


# call the method
create_price_range_column()
```

```python
#@title get_price_binning_frequency( )

def get_price_binning_frequency( ):
  bin_freq_dictionary = { 
      "Low" : 0,
      "Medium" : 0,
      "High" : 0,
      "Very High" : 0
  }
  
  with open("gas_prices_binning.txt") as file:
    for each_line in file:
      linelist=each_line.split(":")
      #print(linelist)
      bin_name = linelist[2]
      bin_name = bin_name[:-1]
      bin_freq_dictionary[bin_name] = bin_freq_dictionary[bin_name] + 1

  # return the value
  return bin_freq_dictionary

bin_dict = get_price_binning_frequency()


print('Printing the Frequency Table')
for key, value in bin_dict.items():
  print(key, '==>', value)

```

```python
#@title get_missing_dates_summary_report() 
import datetime

def get_missing_dates_report():
    file_path = 'gasprices.txt'
    # Read the gas prices from the file
    with open(file_path) as f:
        gas_prices = f.readlines()

    # Extract the dates from the gas prices
    dates = [datetime.datetime.strptime(gp.split(":")[0], "%m-%d-%Y").date() for gp in gas_prices]

    # Create a dictionary to keep track of the missing dates
    missing_dates = {}

    # Loop through all the dates between the first and last date in the file
    start_date = dates[0]
    end_date = dates[-1]
    for date in (start_date + datetime.timedelta(days=x) for x in range((end_date - start_date).days)):
        year_month = date.strftime('%Y %b')

        # Check if the date is missing
        if date not in dates:
            if year_month in missing_dates:
                missing_dates[year_month] += 1
            else:
                missing_dates[year_month] = 1

    # Print the report in a table with 3 columns
    print("Month\tYear\tCount of missing days")
    for year_month, count in missing_dates.items():
        year, month = year_month.split()
        print(f"{month}\t{year}\t{count}")

get_missing_dates_report()
```
